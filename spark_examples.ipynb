```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
```

```python
sc=SparkContext("local[2]","network world count").getOrCreate()
ssc=StreamingContext(sc,1)
```

```python
lines = ssc.socketTextStream("localhost", 9999)
```

```python
word=lines.flatMap(lambda line:line.split(" "))
```

```python
word_map=word.map(lambda word:(word,1))
word_count=word_map.reduceByKey(lambda x,y:x+y)
```

```python
word_count.pprint()
```

```python
ssc.start()
ssc.awaitTermination()
```

```python
data=sc.textFile("alllines.txt")
```

```python
features=data.map(lambda line:line.split(" "))
```

```python
group=features.map(lambda word:(word,1))
group.collect()
```

```python
new_features=data.map(lambda x:(x.split(",")[0],x))
new_features=new_features.map(lambda x:(x[0],x[1]))
```

```python
grouped=new_features.groupByKey(numPartitions=4)
```

```python
grouped.collect()
```

```python
duration=grouped.reduceByKey(lambda x,y:x+y)
```
